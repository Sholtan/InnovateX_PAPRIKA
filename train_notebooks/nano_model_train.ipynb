```python
from ultralytics import YOLO

model = YOLO("yolov8n.pt")
```

```python
model.train(
    data="dataset.yaml",
    epochs=50,
    imgsz=1024,
    batch=-1,
    patience=30,
    lr0=0.001,
    lrf=0.01,
    warmup_epochs=3,
    optimizer="AdamW",
    cos_lr=True,
    augment=True,
    mosaic=0.9,
    close_mosaic=15
)
```
